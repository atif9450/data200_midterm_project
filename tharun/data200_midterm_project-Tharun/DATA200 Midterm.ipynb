```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
emissions_df = pd.read_csv('emissions.csv', encoding='latin-1')
emissions_df.head()
```

```python
nb_emissions = emissions_df[(emissions_df['province'] == 'New Brunswick') & (emissions_df['year'].between(2015, 2019))]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(nb_emissions['year'], nb_emissions['emissions'], marker='o', linestyle='-')
plt.title('Carbon Emissions in New Brunswick (2015-2019)')
plt.xlabel('Year')
plt.ylabel('Carbon Emissions')
plt.grid(True)
plt.xticks(nb_emissions['year'])  # Set x-axis ticks to display all years
plt.tight_layout()
plt.show()
```

```python
nb_parties = emissions_df[emissions_df['province'] == 'Quebec']

# Get unique political parties for each year
parties_by_year = nb_parties.groupby('year')['party'].unique()

# Print the results
for year, parties in parties_by_year.items():
    print(f'Year {year}: {", ".join(parties)}')
```

These two models represent the carbon emissions for New Brunswick and Quebec from 2015 to 2019.
